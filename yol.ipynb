```python
# YOLO Model Test - Basit √á√∂z√ºm
import os
import sys

# Paketleri kontrol et ve y√ºkle
def check_and_install_packages():
    required_packages = {
        'cv2': 'opencv-python',
        'matplotlib': 'matplotlib',
        'numpy': 'numpy',
        'ultralytics': 'ultralytics',
        'PIL': 'Pillow'
    }
    
    missing_packages = []
    
    for module, package in required_packages.items():
        try:
            __import__(module)
        except ImportError:
            missing_packages.append(package)
    
    if missing_packages:
        print(f"Eksik paketler y√ºkleniyor: {missing_packages}")
        for package in missing_packages:
            os.system(f"{sys.executable} -m pip install {package}")
        print("Paketler y√ºklendi! Kernel'i yeniden ba≈ülatƒ±n.")
        return False
    else:
        print("T√ºm paketler mevcut!")
        return True

# Paket kontrol√º
if check_and_install_packages():
    print("Devam edebilirsiniz...")
```

```python
# Gerekli paketleri y√ºkle
import subprocess
import sys

def install_package(package):
    subprocess.check_call([sys.executable, "-m", "pip", "install", package])

packages = ["matplotlib", "opencv-python", "ultralytics", "numpy", "Pillow"]

print("Paketler y√ºkleniyor...")
for package in packages:
    try:
        install_package(package)
        print(f"‚úì {package} y√ºklendi")
    except Exception as e:
        print(f"‚úó {package} y√ºklenemedi: {e}")

print("Paket y√ºkleme tamamlandƒ±!")
```

```python
# YOLO Model Test with Visualization
try:
    import cv2
    import matplotlib.pyplot as plt
    import numpy as np
    from ultralytics import YOLO
    import os
    print("T√ºm k√ºt√ºphaneler ba≈üarƒ±yla y√ºklendi!")
except ImportError as e:
    print(f"K√ºt√ºphane y√ºkleme hatasƒ±: {e}")
    print("L√ºtfen √∂nce paketleri y√ºkleyin.")

# Model ve test g√∂rsel yollarƒ±
model_path = "yol_best"  # Eƒüitilmi≈ü modeliniz
test_image_path = "yol_test.jpeg"  # Test edilecek g√∂rsel

# Dosyalarƒ±n varlƒ±ƒüƒ±nƒ± kontrol et
if not os.path.exists(model_path):
    print(f"HATA: Model dosyasƒ± bulunamadƒ±: {model_path}")
if not os.path.exists(test_image_path):
    print(f"HATA: Test g√∂rseli bulunamadƒ±: {test_image_path}")

try:
    # Modeli y√ºkle
    print("Model y√ºkleniyor...")
    model = YOLO(model_path)
    print("Model ba≈üarƒ±yla y√ºklendi!")
    
    # Test g√∂rselini oku
    print("Test g√∂rseli okunuyor...")
    image = cv2.imread(test_image_path)
    if image is None:
        print("HATA: G√∂rsel okunamadƒ±!")
    else:
        image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        print(f"G√∂rsel boyutu: {image.shape}")
        
        # Modeli test et
        print("Model tahmin yapƒ±yor...")
        results = model(test_image_path)
        
        # Sonu√ßlarƒ± al
        result = results[0]
        boxes = result.boxes
        
        print(f"Tespit edilen nesne sayƒ±sƒ±: {len(boxes)}")
        
        # Matplotlib ile g√∂rselle≈ütirme
        fig, axes = plt.subplots(1, 2, figsize=(15, 7))
        
        # Orijinal g√∂rsel
        axes[0].imshow(image_rgb)
        axes[0].set_title("Orijinal G√∂rsel")
        axes[0].axis('off')
        
        # Tespit sonu√ßlarƒ± ile g√∂rsel
        annotated_image = result.plot()
        annotated_image_rgb = cv2.cvtColor(annotated_image, cv2.COLOR_BGR2RGB)
        
        axes[1].imshow(annotated_image_rgb)
        axes[1].set_title(f"Tespit Sonu√ßlarƒ± ({len(boxes)} nesne)")
        axes[1].axis('off')
        
        plt.tight_layout()
        plt.show()
        
        # Detaylƒ± sonu√ßlarƒ± yazdƒ±r
        if len(boxes) > 0:
            print("\n=== TESPƒ∞T DETAYLARI ===")
            for i, box in enumerate(boxes):
                # Koordinatlar
                x1, y1, x2, y2 = box.xyxy[0].cpu().numpy()
                
                # G√ºven skoru
                confidence = box.conf[0].cpu().numpy()
                
                # Sƒ±nƒ±f
                class_id = int(box.cls[0].cpu().numpy())
                class_name = model.names[class_id] if hasattr(model, 'names') else f"Class_{class_id}"
                
                print(f"\nNesne {i+1}:")
                print(f"  Sƒ±nƒ±f: {class_name}")
                print(f"  G√ºven Skoru: {confidence:.3f}")
                print(f"  Koordinatlar: ({x1:.1f}, {y1:.1f}) - ({x2:.1f}, {y2:.1f})")
                print(f"  Boyut: {x2-x1:.1f} x {y2-y1:.1f} piksel")
        else:
            print("\nHi√ßbir nesne tespit edilmedi.")
        
        print("\nTest tamamlandƒ±!")
        
except Exception as e:
    print(f"Bir hata olu≈ütu: {e}")
    print("L√ºtfen model ve g√∂rsel dosyalarƒ±nƒ±n doƒüru konumda olduƒüundan emin olun.")
```

```python
# YOLO Model Test - Son Versiyon
import cv2
import matplotlib.pyplot as plt
import numpy as np
from ultralytics import YOLO
import os

# Model ve test g√∂rsel yollarƒ±
model_path = "yol_best.pt"  # .pt uzantƒ±lƒ± model dosyasƒ±
test_image_path = "yol_test.jpeg"  # Test edilecek g√∂rsel

print("üöÄ YOLO Model Test Ba≈ülƒ±yor...")
print("="*50)

# Dosyalarƒ±n varlƒ±ƒüƒ±nƒ± kontrol et
if not os.path.exists(model_path):
    print(f"‚ùå HATA: Model dosyasƒ± bulunamadƒ±: {model_path}")
    print("Mevcut dosyalar:")
    for f in os.listdir('.'):
        if 'yol' in f.lower():
            print(f"   - {f}")
    exit()
    
if not os.path.exists(test_image_path):
    print(f"‚ùå HATA: Test g√∂rseli bulunamadƒ±: {test_image_path}")
    exit()

try:
    # Modeli y√ºkle
    print(f"üìÅ Model y√ºkleniyor: {model_path}")
    model = YOLO(model_path)
    print("‚úÖ Model ba≈üarƒ±yla y√ºklendi!")
    
    # Test g√∂rselini oku
    print(f"üñºÔ∏è G√∂rsel okunuyor: {test_image_path}")
    image = cv2.imread(test_image_path)
    if image is None:
        print("‚ùå HATA: G√∂rsel okunamadƒ±!")
        exit()
    else:
        image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        print(f"üìè G√∂rsel boyutu: {image.shape}")
        
        # Model yapƒ±landƒ±rmasƒ± hakkƒ±nda bilgi ver
        if hasattr(model, 'names'):
            print(f"üè∑Ô∏è Model sƒ±nƒ±flarƒ±: {list(model.names.values())}")
        
        # Modeli test et
        print("üîç Model tahmin yapƒ±yor...")
        results = model(test_image_path, conf=0.25, verbose=False)
        
        # Sonu√ßlarƒ± al
        result = results[0]
        boxes = result.boxes
        
        num_detections = len(boxes) if boxes is not None else 0
        print(f"üìä Tespit edilen nesne sayƒ±sƒ±: {num_detections}")
        
        # Matplotlib ile g√∂rselle≈ütirme
        plt.style.use('default')
        fig, axes = plt.subplots(1, 2, figsize=(16, 8))
        fig.suptitle('YOLO Model Test Sonu√ßlarƒ±', fontsize=16, fontweight='bold')
        
        # Orijinal g√∂rsel
        axes[0].imshow(image_rgb)
        axes[0].set_title("Orijinal G√∂rsel", fontsize=14, fontweight='bold')
        axes[0].axis('off')
        
        # Tespit sonu√ßlarƒ± ile g√∂rsel
        annotated_image = result.plot()
        annotated_image_rgb = cv2.cvtColor(annotated_image, cv2.COLOR_BGR2RGB)
        
        axes[1].imshow(annotated_image_rgb)
        axes[1].set_title(f"Tespit Sonu√ßlarƒ± ({num_detections} nesne)", fontsize=14, fontweight='bold')
        axes[1].axis('off')
        
        plt.tight_layout()
        plt.show()
        
        # Detaylƒ± sonu√ßlarƒ± yazdƒ±r
        if boxes is not None and len(boxes) > 0:
            print("\n" + "="*60)
            print("üéØ TESPƒ∞T DETAYLARI")
            print("="*60)
            
            total_area = 0
            for i, box in enumerate(boxes):
                # Koordinatlar
                x1, y1, x2, y2 = box.xyxy[0].cpu().numpy()
                
                # G√ºven skoru
                confidence = box.conf[0].cpu().numpy()
                
                # Sƒ±nƒ±f
                class_id = int(box.cls[0].cpu().numpy())
                class_name = model.names[class_id] if hasattr(model, 'names') else f"Class_{class_id}"
                
                # Alan hesapla
                area = (x2-x1)*(y2-y1)
                total_area += area
                
                print(f"\nüîç Nesne {i+1}:")
                print(f"   üìã Sƒ±nƒ±f: {class_name}")
                print(f"   üéØ G√ºven Skoru: {confidence:.3f} ({confidence*100:.1f}%)")
                print(f"   üìç Koordinatlar: ({x1:.1f}, {y1:.1f}) ‚Üí ({x2:.1f}, {y2:.1f})")
                print(f"   üìè Boyut: {x2-x1:.1f} √ó {y2-y1:.1f} piksel")
                print(f"   üìê Alan: {area:.0f} piksel¬≤")
                
                # G√∂rsel i√ßindeki konumu
                center_x = (x1 + x2) / 2
                center_y = (y1 + y2) / 2
                img_center_x = image.shape[1] / 2
                img_center_y = image.shape[0] / 2
                
                if center_x < img_center_x / 2:
                    h_pos = "sol"
                elif center_x > img_center_x * 1.5:
                    h_pos = "saƒü"
                else:
                    h_pos = "orta"
                    
                if center_y < img_center_y / 2:
                    v_pos = "√ºst"
                elif center_y > img_center_y * 1.5:
                    v_pos = "alt"
                else:
                    v_pos = "orta"
                    
                print(f"   üìç Konum: {v_pos}-{h_pos}")
                
            print(f"\n� √ñZET ƒ∞STATƒ∞STƒ∞KLER:")
            print(f"   ‚Ä¢ Toplam tespit: {len(boxes)} nesne")
            print(f"   ‚Ä¢ Toplam alan: {total_area:.0f} piksel¬≤")
            print(f"   ‚Ä¢ Ortalama g√ºven: {np.mean([box.conf[0].cpu().numpy() for box in boxes]):.3f}")
            print(f"   ‚Ä¢ En y√ºksek g√ºven: {max([box.conf[0].cpu().numpy() for box in boxes]):.3f}")
            print(f"   ‚Ä¢ En d√º≈ü√ºk g√ºven: {min([box.conf[0].cpu().numpy() for box in boxes]):.3f}")
            
        else:
            print(f"\n‚ùå Hi√ßbir nesne tespit edilmedi.")
            print("\nüí° ƒ∞PU√áLARI:")
            print("   üîß G√ºven e≈üiƒüini d√º≈ü√ºrmeyi deneyin (conf < 0.25)")
            print("   üñºÔ∏è G√∂rsel kalitesini kontrol edin")
            print("   üéì Model eƒüitiminin yeterli olup olmadƒ±ƒüƒ±nƒ± kontrol edin")
            print("   üìä Veri setinin √ße≈üitliliƒüini artƒ±rƒ±n")
        
        print(f"\n‚úÖ Test ba≈üarƒ±yla tamamlandƒ±!")
        print("="*50)
        
except Exception as e:
    print(f"\n‚ùå Bir hata olu≈ütu: {e}")
    print("\nüîß OLASI √á√ñZ√úMLER:")
    print("   ‚Ä¢ Model dosyasƒ±nƒ±n PyTorch (.pt) formatƒ±nda olduƒüundan emin olun")
    print("   ‚Ä¢ Model dosyasƒ±nƒ±n bozuk olmadƒ±ƒüƒ±nƒ± kontrol edin")
    print("   ‚Ä¢ Ultralytics s√ºr√ºm√ºn√ºz√º g√ºncelleyin: pip install -U ultralytics")
    print("   ‚Ä¢ Modeli yeniden eƒüitip kaydetmeyi deneyin")
    import traceback
    print(f"\nüêõ Detaylƒ± hata: {traceback.format_exc()}")
```
