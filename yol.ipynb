```python
# YOLO Model Test - Basit Ã‡Ã¶zÃ¼m
import os
import sys

# Paketleri kontrol et ve yÃ¼kle
def check_and_install_packages():
    required_packages = {
        'cv2': 'opencv-python',
        'matplotlib': 'matplotlib',
        'numpy': 'numpy',
        'ultralytics': 'ultralytics',
        'PIL': 'Pillow'
    }
    
    missing_packages = []
    
    for module, package in required_packages.items():
        try:
            __import__(module)
        except ImportError:
            missing_packages.append(package)
    
    if missing_packages:
        print(f"Eksik paketler yÃ¼kleniyor: {missing_packages}")
        for package in missing_packages:
            os.system(f"{sys.executable} -m pip install {package}")
        print("Paketler yÃ¼klendi! Kernel'i yeniden baÅŸlatÄ±n.")
        return False
    else:
        print("TÃ¼m paketler mevcut!")
        return True

# Paket kontrolÃ¼
if check_and_install_packages():
    print("Devam edebilirsiniz...")
```

```python
# Gerekli paketleri yÃ¼kle
import subprocess
import sys

def install_package(package):
    subprocess.check_call([sys.executable, "-m", "pip", "install", package])

packages = ["matplotlib", "opencv-python", "ultralytics", "numpy", "Pillow"]

print("Paketler yÃ¼kleniyor...")
for package in packages:
    try:
        install_package(package)
        print(f"âœ“ {package} yÃ¼klendi")
    except Exception as e:
        print(f"âœ— {package} yÃ¼klenemedi: {e}")

print("Paket yÃ¼kleme tamamlandÄ±!")
```

```python
# YOLO Model Test with Visualization
try:
    import cv2
    import matplotlib.pyplot as plt
    import numpy as np
    from ultralytics import YOLO
    import os
    print("TÃ¼m kÃ¼tÃ¼phaneler baÅŸarÄ±yla yÃ¼klendi!")
except ImportError as e:
    print(f"KÃ¼tÃ¼phane yÃ¼kleme hatasÄ±: {e}")
    print("LÃ¼tfen Ã¶nce paketleri yÃ¼kleyin.")

# Model ve test gÃ¶rsel yollarÄ±
model_path = "yol_best"  # EÄŸitilmiÅŸ modeliniz
test_image_path = "yol_test.jpeg"  # Test edilecek gÃ¶rsel

# DosyalarÄ±n varlÄ±ÄŸÄ±nÄ± kontrol et
if not os.path.exists(model_path):
    print(f"HATA: Model dosyasÄ± bulunamadÄ±: {model_path}")
if not os.path.exists(test_image_path):
    print(f"HATA: Test gÃ¶rseli bulunamadÄ±: {test_image_path}")

try:
    # Modeli yÃ¼kle
    print("Model yÃ¼kleniyor...")
    model = YOLO(model_path)
    print("Model baÅŸarÄ±yla yÃ¼klendi!")
    
    # Test gÃ¶rselini oku
    print("Test gÃ¶rseli okunuyor...")
    image = cv2.imread(test_image_path)
    if image is None:
        print("HATA: GÃ¶rsel okunamadÄ±!")
    else:
        image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        print(f"GÃ¶rsel boyutu: {image.shape}")
        
        # Modeli test et
        print("Model tahmin yapÄ±yor...")
        results = model(test_image_path)
        
        # SonuÃ§larÄ± al
        result = results[0]
        boxes = result.boxes
        
        print(f"Tespit edilen nesne sayÄ±sÄ±: {len(boxes)}")
        
        # Matplotlib ile gÃ¶rselleÅŸtirme
        fig, axes = plt.subplots(1, 2, figsize=(15, 7))
        
        # Orijinal gÃ¶rsel
        axes[0].imshow(image_rgb)
        axes[0].set_title("Orijinal GÃ¶rsel")
        axes[0].axis('off')
        
        # Tespit sonuÃ§larÄ± ile gÃ¶rsel
        annotated_image = result.plot()
        annotated_image_rgb = cv2.cvtColor(annotated_image, cv2.COLOR_BGR2RGB)
        
        axes[1].imshow(annotated_image_rgb)
        axes[1].set_title(f"Tespit SonuÃ§larÄ± ({len(boxes)} nesne)")
        axes[1].axis('off')
        
        plt.tight_layout()
        plt.show()
        
        # DetaylÄ± sonuÃ§larÄ± yazdÄ±r
        if len(boxes) > 0:
            print("\n=== TESPÄ°T DETAYLARI ===")
            for i, box in enumerate(boxes):
                # Koordinatlar
                x1, y1, x2, y2 = box.xyxy[0].cpu().numpy()
                
                # GÃ¼ven skoru
                confidence = box.conf[0].cpu().numpy()
                
                # SÄ±nÄ±f
                class_id = int(box.cls[0].cpu().numpy())
                class_name = model.names[class_id] if hasattr(model, 'names') else f"Class_{class_id}"
                
                print(f"\nNesne {i+1}:")
                print(f"  SÄ±nÄ±f: {class_name}")
                print(f"  GÃ¼ven Skoru: {confidence:.3f}")
                print(f"  Koordinatlar: ({x1:.1f}, {y1:.1f}) - ({x2:.1f}, {y2:.1f})")
                print(f"  Boyut: {x2-x1:.1f} x {y2-y1:.1f} piksel")
        else:
            print("\nHiÃ§bir nesne tespit edilmedi.")
        
        print("\nTest tamamlandÄ±!")
        
except Exception as e:
    print(f"Bir hata oluÅŸtu: {e}")
    print("LÃ¼tfen model ve gÃ¶rsel dosyalarÄ±nÄ±n doÄŸru konumda olduÄŸundan emin olun.")
```

```python
# YOLO Model Test - Son Versiyon
import cv2
import matplotlib.pyplot as plt
import numpy as np
from ultralytics import YOLO
import os

# Model ve test gÃ¶rsel yollarÄ±
model_path = "yol_best.pt"  # .pt uzantÄ±lÄ± model dosyasÄ±
test_image_path = "yol_test.jpeg"  # Test edilecek gÃ¶rsel

print("ğŸš€ YOLO Model Test BaÅŸlÄ±yor...")
print("="*50)

# DosyalarÄ±n varlÄ±ÄŸÄ±nÄ± kontrol et
if not os.path.exists(model_path):
    print(f"âŒ HATA: Model dosyasÄ± bulunamadÄ±: {model_path}")
    print("Mevcut dosyalar:")
    for f in os.listdir('.'):
        if 'yol' in f.lower():
            print(f"   - {f}")
    exit()
    
if not os.path.exists(test_image_path):
    print(f"âŒ HATA: Test gÃ¶rseli bulunamadÄ±: {test_image_path}")
    exit()

try:
    # Modeli yÃ¼kle
    print(f"ğŸ“ Model yÃ¼kleniyor: {model_path}")
    model = YOLO(model_path)
    print("âœ… Model baÅŸarÄ±yla yÃ¼klendi!")
    
    # Test gÃ¶rselini oku
    print(f"ğŸ–¼ï¸ GÃ¶rsel okunuyor: {test_image_path}")
    image = cv2.imread(test_image_path)
    if image is None:
        print("âŒ HATA: GÃ¶rsel okunamadÄ±!")
        exit()
    else:
        image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        print(f"ğŸ“ GÃ¶rsel boyutu: {image.shape}")
        
        # Model yapÄ±landÄ±rmasÄ± hakkÄ±nda bilgi ver
        if hasattr(model, 'names'):
            print(f"ğŸ·ï¸ Model sÄ±nÄ±flarÄ±: {list(model.names.values())}")
        
        # Modeli test et
        print("ğŸ” Model tahmin yapÄ±yor...")
        results = model(test_image_path, conf=0.25, verbose=False)
        
        # SonuÃ§larÄ± al
        result = results[0]
        boxes = result.boxes
        
        num_detections = len(boxes) if boxes is not None else 0
        print(f"ğŸ“Š Tespit edilen nesne sayÄ±sÄ±: {num_detections}")
        
        # Matplotlib ile gÃ¶rselleÅŸtirme
        plt.style.use('default')
        fig, axes = plt.subplots(1, 2, figsize=(16, 8))
        fig.suptitle('YOLO Model Test SonuÃ§larÄ±', fontsize=16, fontweight='bold')
        
        # Orijinal gÃ¶rsel
        axes[0].imshow(image_rgb)
        axes[0].set_title("Orijinal GÃ¶rsel", fontsize=14, fontweight='bold')
        axes[0].axis('off')
        
        # Tespit sonuÃ§larÄ± ile gÃ¶rsel
        annotated_image = result.plot()
        annotated_image_rgb = cv2.cvtColor(annotated_image, cv2.COLOR_BGR2RGB)
        
        axes[1].imshow(annotated_image_rgb)
        axes[1].set_title(f"Tespit SonuÃ§larÄ± ({num_detections} nesne)", fontsize=14, fontweight='bold')
        axes[1].axis('off')
        
        plt.tight_layout()
        plt.show()
        
        # DetaylÄ± sonuÃ§larÄ± yazdÄ±r
        if boxes is not None and len(boxes) > 0:
            print("\n" + "="*60)
            print("ğŸ¯ TESPÄ°T DETAYLARI")
            print("="*60)
            
            total_area = 0
            for i, box in enumerate(boxes):
                # Koordinatlar
                x1, y1, x2, y2 = box.xyxy[0].cpu().numpy()
                
                # GÃ¼ven skoru
                confidence = box.conf[0].cpu().numpy()
                
                # SÄ±nÄ±f
                class_id = int(box.cls[0].cpu().numpy())
                class_name = model.names[class_id] if hasattr(model, 'names') else f"Class_{class_id}"
                
                # Alan hesapla
                area = (x2-x1)*(y2-y1)
                total_area += area
                
                print(f"\nğŸ” Nesne {i+1}:")
                print(f"   ğŸ“‹ SÄ±nÄ±f: {class_name}")
                print(f"   ğŸ¯ GÃ¼ven Skoru: {confidence:.3f} ({confidence*100:.1f}%)")
                print(f"   ğŸ“ Koordinatlar: ({x1:.1f}, {y1:.1f}) â†’ ({x2:.1f}, {y2:.1f})")
                print(f"   ğŸ“ Boyut: {x2-x1:.1f} Ã— {y2-y1:.1f} piksel")
                print(f"   ğŸ“ Alan: {area:.0f} pikselÂ²")
                
                # GÃ¶rsel iÃ§indeki konumu
                center_x = (x1 + x2) / 2
                center_y = (y1 + y2) / 2
                img_center_x = image.shape[1] / 2
                img_center_y = image.shape[0] / 2
                
                if center_x < img_center_x / 2:
                    h_pos = "sol"
                elif center_x > img_center_x * 1.5:
                    h_pos = "saÄŸ"
                else:
                    h_pos = "orta"
                    
                if center_y < img_center_y / 2:
                    v_pos = "Ã¼st"
                elif center_y > img_center_y * 1.5:
                    v_pos = "alt"
                else:
                    v_pos = "orta"
                    
                print(f"   ğŸ“ Konum: {v_pos}-{h_pos}")
                
            print(f"\n� Ã–ZET Ä°STATÄ°STÄ°KLER:")
            print(f"   â€¢ Toplam tespit: {len(boxes)} nesne")
            print(f"   â€¢ Toplam alan: {total_area:.0f} pikselÂ²")
            print(f"   â€¢ Ortalama gÃ¼ven: {np.mean([box.conf[0].cpu().numpy() for box in boxes]):.3f}")
            print(f"   â€¢ En yÃ¼ksek gÃ¼ven: {max([box.conf[0].cpu().numpy() for box in boxes]):.3f}")
            print(f"   â€¢ En dÃ¼ÅŸÃ¼k gÃ¼ven: {min([box.conf[0].cpu().numpy() for box in boxes]):.3f}")
            
        else:
            print(f"\nâŒ HiÃ§bir nesne tespit edilmedi.")
            print("\nğŸ’¡ Ä°PUÃ‡LARI:")
            print("   ğŸ”§ GÃ¼ven eÅŸiÄŸini dÃ¼ÅŸÃ¼rmeyi deneyin (conf < 0.25)")
            print("   ğŸ–¼ï¸ GÃ¶rsel kalitesini kontrol edin")
            print("   ğŸ“ Model eÄŸitiminin yeterli olup olmadÄ±ÄŸÄ±nÄ± kontrol edin")
            print("   ğŸ“Š Veri setinin Ã§eÅŸitliliÄŸini artÄ±rÄ±n")
        
        print(f"\nâœ… Test baÅŸarÄ±yla tamamlandÄ±!")
        print("="*50)
        
except Exception as e:
    print(f"\nâŒ Bir hata oluÅŸtu: {e}")
    print("\nğŸ”§ OLASI Ã‡Ã–ZÃœMLER:")
    print("   â€¢ Model dosyasÄ±nÄ±n PyTorch (.pt) formatÄ±nda olduÄŸundan emin olun")
    print("   â€¢ Model dosyasÄ±nÄ±n bozuk olmadÄ±ÄŸÄ±nÄ± kontrol edin")
    print("   â€¢ Ultralytics sÃ¼rÃ¼mÃ¼nÃ¼zÃ¼ gÃ¼ncelleyin: pip install -U ultralytics")
    print("   â€¢ Modeli yeniden eÄŸitip kaydetmeyi deneyin")
    import traceback
    print(f"\nğŸ› DetaylÄ± hata: {traceback.format_exc()}")
```
